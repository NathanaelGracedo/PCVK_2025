**1. Dengan menggunakan figsize, perhatikan apakah ukuran image pixelnya juga berubah?**

````python
img_tugas = cv.imread('/content/drive/MyDrive/PCVK_img/i_want_to_ask_ah_dog.jpg')
plt.imshow(cv.cvtColor(img_tugas, cv.COLOR_BGR2RGB))
plt.title('Gambar Asli')

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(442)
plt.imshow(cv.cvtColor(img_tugas, cv.COLOR_BGR2RGB))
plt.title('figsize=(7, 7)')
plt.show()

print("Setelah dilakukan figsize hanya mengubah ukuran tampilan figure, bukan ukuran piksel data gambar.")
````

**2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!**

````python
img_rgb = cv.cvtColor(img_tugas, cv.COLOR_BGR2RGB)
R, G, B = img_rgb[:,:,0], img_rgb[:,:,1], img_rgb[:,:,2]
zeros = np.zeros_like(R)

# Red-Blue
rb = np.dstack([R, zeros, B])
plt.figure()
plt.imshow(rb)
plt.title('Red-Blue')
plt.show()

# Green-Blue
gb = np.dstack([zeros, G, B])
plt.figure()
plt.imshow(gb)
plt.title('Green-Blue')
plt.show()
````
