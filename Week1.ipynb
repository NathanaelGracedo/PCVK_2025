# **Praktikum D2**

Step 1: Import library numpy, pandas, cv2, skimage, PIL, *matplotlib*

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

Step 2: Membaca dan menampilkan gambar

```python
urls = ["https://i.pinimg.com/736x/61/46/09/614609c31f53e9ced28b7df46642eb17.jpg",
        "https://i.pinimg.com/736x/87/91/51/879151bff9e6be0f691a7c64f9be0221.jpg",
        "https://i.pinimg.com/736x/03/8b/9b/038b9b255a176dfd805b67cbf9585c0e.jpg"]

for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print("\n")
```

Step 3: Melihat ukuran gambar

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi imageL: tinggi x lebar =", tinggi, " x ",lebar)
cv2_imshow(image_2)
```

Step 4: Mengakses pixel dengan memberi garis diagonal menyilang

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal pada tengah image
for y in range (lebar):
  image_3[int((tinggi)/2), y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

3. **Pada uji coba langkah ke-9 terdapat potongan kode program sebagai berikut :**

image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)

**Apa kegunaan kode program tersebut?dan apa pengaruhnya jika tidak
dilakukan?**

jawab:

Kode program tersebut berfungsi untuk mengecilkan gambar menjadi 50% pada lebar (fx=0.5) dan tinggi (fy=0.5). Karena dsize=(0,0), OpenCV mengabaikan ukuran absolut dan memakai faktor skala. Hasilnya resolusi turun jadi 25% jumlah piksel awal, memori lebih hemat, komputasi lebih cepat, dan lebih mudah ditampilkan berdampingan (hconcat) di Colab tanpa terlalu lebar.

Jika tidak dilakukan output gambar tetap pada resolusi asli sehingga hasil gabungan dua gambar bisa terlalu besar untuk tampilan notebook.

4. **Apakah kegunaan kode [255,255,255] ? Jelaskan!**

jawab:
Pada rentang piksel pada RGB/BGR 255,255,255 adalah nilai warna putih, yang dimana tiap angka 255 mempresentasikan warna R (red), G (Green), dan B (Blue) dengan rentang 0-255. Jadi kegunaan kode tersebut adalah untuk mengubah
warna pixel yang sudah di tentukan menjadi warna putih.

5. **Jelaskan keterkaitan antara pixel dan juga resolusi gambar yang tinggi ataupun rendah!**

jawab:

Tiap pixel menyimpan nilai warna/intensitas. Resolusi adalah jumlah total pixel pada gambar (lebar × tinggi). Semakin tinggi resolusi, semakin banyak pixel yang dimiliki, sehingga detail yang dapat direkam/ditampilkan lebih kaya, gambar tampak tajam saat diperbesar atau dicetak besar, tetapi ukuran file dan beban komputasi juga meningkat. Sebaliknya, resolusi rendah berarti jumlah pixel sedikit: proses lebih ringan dan file lebih kecil, namun detail berkurang dan gambar cepat tampak pecah saat di-zoom.
